```python
import numpy as np
```

```python
np.version.version
```

```python
x = np.random.rand(2048, 2048).astype(dtype=np.float32) / 5.0
```

```python
def bench_func(x):
    for i in range(3):
        x = (np.matmul(x,x)+x)/1000.0
    return x
```

```python
%timeit bench_func(x)
```

```python
import torch
```

```python
torch.__version__
```

```python
xt = torch.tensor(x)
```

```python
def bench_func_t(x):
    for i in range(3):
        x = (torch.matmul(x,x)+x)/1000.0
    return x
```

```python
%timeit bench_func_t(xt)
```

```python
bench_func_tc = torch.compile(bench_func_t)
```

```python
%timeit bench_func_tc(xt)
```
